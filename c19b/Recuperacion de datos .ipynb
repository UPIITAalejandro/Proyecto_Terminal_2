# Interpolacion de datos 

## Introduccion

Como situacion actual se tiene que los documentos descargados de esta insitucion cuentan con un periodo vacio entre un a√±o y otro. Por ello, se recurrira a un metodo de interpolacion para poder solucionar este problema para de esta manera no contar con un ajuste no deseado en el modelo

## Objetivo

determinar este valor faltante por medio de una interpolacion elaborada en los DataFrame.


Ademas se planea normalizar los datos para de esta manera en futuras actividades realizar alguna vase de datos


```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df_2008 = pd.read_csv("2008.csv")
df_2010 = pd.read_csv("2010.csv")
df_2012 = pd.read_csv("2012.csv")
df_2014 = pd.read_csv("2014.csv")
df_2016 = pd.read_csv("2016.csv")
df_2018 = pd.read_csv("2018.csv")
```

```python
# Creacion del diccionario para datos faltantes
diccionario_anios_faltantes = {
    'estado_1':df_2008['estado_1'], 
    'anio':list(np.nan for i in range(0,32)), 
    'porc_pob_afiliada_seguro_popular':list(np.nan for i in range(0,32)),
    'miles_perso_afiliadas_seguro_popular':list(np.nan for i in range(0,32)), 
    'porc_pob_afiliada_imss':list(np.nan for i in range(0,32)),
    'miles_perso_afilidas_imss':list(np.nan for i in range(0,32)), 
    'porc_pob_afiliada_issste':list(np.nan for i in range(0,32)),
    'miles_perso_afilidas_issste':list(np.nan for i in range(0,32)), 
    'porc_pob_afiliada_issste_estatal':list(np.nan for i in range(0,32)),
    'miles_perso_afilidas_issste_estatal':list(np.nan for i in range(0,32)),
    'porc_pob_afiliada_pemex_defensa_o_marina':list(np.nan for i in range(0,32)),
    'miles_perso_afiliadas_pemex_defensa_o_marina':list(np.nan for i in range(0,32)),
    'porc_pob_seguro_privado_gastos_med':list(np.nan for i in range(0,32)),
    'miles_perso_seguro_privado_gastos_med':list(np.nan for i in range(0,32)),
    'porc_pob_acces_servicios_med_seg_social_indirect':list(np.nan for i in range(0,32)),
    'miles_perso_acces_servicios_med_seg_social_indirect':list(np.nan for i in range(0,32)),
    'porc_pob_afiliada_inst_medica_dist_las_ant':list(np.nan for i in range(0,32)),
    'miles_perso_afiliadas_inst_medica_dist_las_ant':list(np.nan for i in range(0,32))
}
```

```python
# Creacion de diccionario con datos faltantes

df_2009 = pd.DataFrame(diccionario_anios_faltantes)
df_2009['anio'] = 2009

df_2011 = pd.DataFrame(diccionario_anios_faltantes)
df_2011['anio'] = 2011

df_2013 = pd.DataFrame(diccionario_anios_faltantes)
df_2013['anio'] = 2013

df_2015 = pd.DataFrame(diccionario_anios_faltantes)
df_2015['anio'] = 2015

df_2017 = pd.DataFrame(diccionario_anios_faltantes)
df_2017['anio'] = 2017
```

```python
df_general = pd.DataFrame()
df_general
```

```python
for i in range(2,len(list(df_2008.columns))):
    
    #Recorrer para interpolacion 
    aux_2008 = df_2008.iloc[:,lambda df_2008:[i]]
    aux_2009 = df_2009.iloc[:,lambda df_2008:[i]]
    aux_2010 = df_2010.iloc[:,lambda df_2008:[i]]
    aux_2011 = df_2011.iloc[:,lambda df_2008:[i]]
    aux_2012 = df_2012.iloc[:,lambda df_2008:[i]]
    aux_2013 = df_2013.iloc[:,lambda df_2008:[i]]
    aux_2014 = df_2014.iloc[:,lambda df_2008:[i]]
    aux_2015 = df_2015.iloc[:,lambda df_2008:[i]]
    aux_2016 = df_2016.iloc[:,lambda df_2008:[i]]
    aux_2017 = df_2017.iloc[:,lambda df_2008:[i]]
    aux_2018 = df_2018.iloc[:,lambda df_2008:[i]]

    df_general_aux = pd.concat([aux_2008,aux_2009,aux_2010,aux_2011,aux_2012,aux_2013,aux_2014,aux_2015,aux_2016,aux_2017,aux_2018],axis=1)
    df_general_aux = df_general_aux.interpolate(method='linear',axis = 1).copy()
    
    lista_nombre_columnas = [list(aux_2008.columns)[0]+"_"+str(df_2008['anio'][0]),
        list(aux_2009.columns)[0]+"_"+str(df_2009['anio'][0]),list(aux_2010.columns)[0]+"_"+str(df_2010['anio'][0]),
        list(aux_2011.columns)[0]+"_"+str(df_2011['anio'][0]),list(aux_2012.columns)[0]+"_"+str(df_2012['anio'][0]),
        list(aux_2013.columns)[0]+"_"+str(df_2013['anio'][0]),list(aux_2014.columns)[0]+"_"+str(df_2014['anio'][0]),
        list(aux_2015.columns)[0]+"_"+str(df_2015['anio'][0]),list(aux_2016.columns)[0]+"_"+str(df_2016['anio'][0]),
        list(aux_2017.columns)[0]+"_"+str(df_2017['anio'][0]),list(aux_2018.columns)[0]+"_"+str(df_2018['anio'][0])]
    
    df_general_aux.columns = lista_nombre_columnas 
    
    
    df_general=pd.concat([df_general,df_general_aux],axis=1)
```

```python
df_nombre = pd.concat([df_2008.iloc[:,[0]],df_2008.iloc[:,[0]]],axis = 1)
df_nombre.columns = ['Id','Estado']
df_nombre['Estado'] = df_nombre['Estado'].str.replace('_',' ')
df_nombre['Estado'] = df_nombre['Estado'].str.title()

df_nombre['Id'] = df_nombre['Estado']

diccionario_id = {
 'Aguascalientes':1,
 'Baja California':2,
 'Baja California Sur':3,
 'Campeche':4,
 'Chiapas':5,
 'Chihuahua':6,
 'Ciudad De Mexico':7,
 'Coahuila':8,
 'Colima':9,
 'Durango':10,
 'Guanajuato':11,
 'Guerrero':12,
 'Hidalgo':13,
 'Jalisco':14,
 'Mexico':15,
 'Michoacan':16,
 'Morelos':17,
 'Nayarit':18,
 'Nuevo Leon':19,
 'Oaxaca':20,
 'Puebla':21,
 'Queretaro':22,
 'Quintana Roo':23,
 'San Luis Potosi':24,
 'Sinaloa':25,
 'Sonora':26,
 'Tabasco':27,
 'Tamaulipas':28,
 'Tlaxcala':29,
 'Veracruz':30,
 'Yucatan':31,
 'Zacatecas':32
}


df_nombre['Id'] = df_nombre['Id'].replace(diccionario_id)
```

```python
df_unido = pd.concat([df_nombre,df_general],axis=1)
```

```python
df_final = df_unido.sort_values(by=['Id'])
```
