```python
# Creacion del diccionario para datos faltantes
diccionario_anios_faltantes = {
    'Entidadfederativa':df_2010['Entidadfederativa'], 
    'Periodo':list(np.nan for i in range(0,32)),
    'Hogarescongastoensistemasalternativossalud_segúndecildeingresos':list(np.nan for i in range(0,32)),
    'Hogaressegúndecilesdeingreso':list(np.nan for i in range(0,32)),
    'Hogarescongastoensistemasalternativossaluddecilesdeingreso':list(np.nan for i in range(0,32))
}
```

```python
# Creacion de diccionario con datos faltantes

df_2011 = pd.DataFrame(diccionario_anios_faltantes)
df_2011['Periodo'] = 2011

df_2013 = pd.DataFrame(diccionario_anios_faltantes)
df_2013['Periodo'] = 2013

df_2015 = pd.DataFrame(diccionario_anios_faltantes)
df_2015['Periodo'] = 2015
```

```python
df_general = pd.DataFrame()
```

```python
for i in range(2,len(list(df_2010.columns))):
    
    #Recorrer para interpolacion 
    aux_2010 = df_2010.iloc[:,lambda df_2010:[i]]
    aux_2011 = df_2011.iloc[:,lambda df_2010:[i]]
    aux_2012 = df_2012.iloc[:,lambda df_2010:[i]]
    aux_2013 = df_2013.iloc[:,lambda df_2010:[i]]
    aux_2014 = df_2014.iloc[:,lambda df_2010:[i]]
    aux_2015 = df_2015.iloc[:,lambda df_2010:[i]]
    aux_2016 = df_2016.iloc[:,lambda df_2010:[i]]

    df_general_aux = pd.concat([aux_2010,aux_2011,aux_2012,aux_2013,aux_2014,aux_2015,aux_2016],axis=1)
    df_general_aux = df_general_aux.interpolate(method='linear',axis = 1).copy()
    
    lista_nombre_columnas = [list(aux_2010.columns)[0]+"_"+str(df_2010['Periodo'][0]),
        list(aux_2011.columns)[0]+"_"+str(df_2011['Periodo'][0]),list(aux_2012.columns)[0]+"_"+str(df_2012['Periodo'][0]),
        list(aux_2013.columns)[0]+"_"+str(df_2013['Periodo'][0]),list(aux_2014.columns)[0]+"_"+str(df_2014['Periodo'][0]),
        list(aux_2015.columns)[0]+"_"+str(df_2015['Periodo'][0]),list(aux_2016.columns)[0]+"_"+str(df_2016['Periodo'][0])]
    
    df_general_aux.columns = lista_nombre_columnas 
    
    
    df_general=pd.concat([df_general,df_general_aux],axis=1)
```
