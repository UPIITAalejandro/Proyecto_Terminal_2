```python
# Dropear columnas 

dfn_2010 = df_2010.drop(columns=['0_6_MESES_OBESIDAD','6_23_MESES_OBESIDAD']).copy()
dfn_2011 = df_2011.drop(columns=['0_6_MESES_OBESIDAD','6_23_MESES_OBESIDAD']).copy()
dfn_2012 = df_2012.drop(columns=['0_6_MESES_OBESIDAD','6_23_MESES_OBESIDAD']).copy()
dfn_2013 = df_2013.drop(columns=['0_6_MESES_OBESIDAD','6_23_MESES_OBESIDAD']).copy()
dfn_2014 = df_2014.drop(columns=['0_1_ANO_OBESIDAD', '1_2_ANIOS_OBESIDAD']).copy()
dfn_2015 = df_2015.drop(columns=['0_1_ANO_OBESIDAD', '1_2_ANIOS_OBESIDAD']).copy()
dfn_2016 = df_2016.drop(columns=['0_6_MESES_OBESIDAD', '6_11_MESES_OBESIDAD',
       '12_23_MESES_OBESIDAD']).copy()
dfn_2017 = df_2017.drop(columns=['0_6_MESES_OBESIDAD', '6_11_MESES_OBESIDAD',
       '12_23_MESES_OBESIDAD']).copy()
dfn_2018 = df_2018.drop(columns=['0_6_MESES_OBESIDAD', '6_11_MESES_OBESIDAD',
       '12_23_MESES_OBESIDAD']).copy()
```

```python
df_general['ENTIDAD'] = df_general['ENTIDAD'].str.lower()
df_general['ENTIDAD'] = df_general['ENTIDAD'].str.title()

df_general.ENTIDAD.value_counts()
```

```python
df_general['ENTIDAD']=df_general['ENTIDAD'].str.replace('Distrito Federal','Ciudad De Mexico')
df_general['ENTIDAD']=df_general['ENTIDAD'].str.replace('Veracruz De Ignacio De La Llave','Veracruz')
df_general['ENTIDAD']=df_general['ENTIDAD'].str.replace('Michoacan De Ocampo','Michoacan')
df_general['ENTIDAD']=df_general['ENTIDAD'].str.replace('Queretaro Arteaga','Queretaro')
df_general['ENTIDAD']=df_general['ENTIDAD'].str.replace('Coahuila De Zaragoza','Coahuila')
```
